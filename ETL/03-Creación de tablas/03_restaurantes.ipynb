```python
import pandas as pd
import json

from fuzzywuzzy import process
```

### Para este codigo, debemos haber ejecutado anteriormente *01_restaurantes* y *estados*, ya que trabajaremos con los archivos obtenidos

```python
# Especifica la ruta del archivo JSON
ruta_archivo = 'restaurantes_A.json'
# Lee el archivo JSON y carga los datos
with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
    # Carga el archivo JSON
    objetos_json = [json.loads(line) for line in archivo]
# Convierte los datos en un DataFrame
df_restaurantes = pd.DataFrame(objetos_json)
```

```python
df_restaurantes.head(5)
```

```python
# Lee el archivo xlsx y carga los datos en un DataFrame
df_ciudades = pd.read_excel('cities_florida.xlsx')

lista_ciudades = df_ciudades['city'].tolist()

df_ciudades = df_ciudades.reset_index(drop=True).reset_index()
# Florida es el estado con id nro 12
df_ciudades['id_estado'] = '12'

df_ciudades.columns = ['id_ciudad', 'ciudad', 'id_estado']
df_ciudades = df_ciudades[['id_ciudad', 'id_estado', 'ciudad']]
```

```python
df_ciudades
```

```python
lista_ciudades
```

```python
# Buscaremos sobre las ciudades de nuestro dataframe, las mejores coincidencias con las ciudades de florida que obtuvimos de internet.
for index, row in df_restaurantes.iterrows():
    mejor_coincidencia, _ = process.extractOne(row['ciudad'], lista_ciudades)
    df_restaurantes.loc[index, 'ciudad_corregida'] = mejor_coincidencia
```

```python
df_restaurantes.head(3)
```

```python
# Unimos las tablas para obtener el id de las ciudades
df_restaurantes = pd.merge(df_restaurantes, df_ciudades, left_on='ciudad_corregida', right_on='ciudad', how='left')
```

```python
df_restaurantes.head(3)
```

```python
# Eliminamos todas las tablas intermedias que no serviran mas
df_restaurantes = df_restaurantes.drop(['categorias', 'ciudad_x','ciudad_corregida', 'id_estado','ciudad_y'], axis=1)
# Cambiamos nombres y ordenamos
df_restaurantes.columns = ['id_restaurante', 'nombre', 'identificador_google', 'latitud', 'longitud', 'avg_rating', 'reviews_count', 'servicios', 'estado', 'direccion', 'identificador_yelp', 'id_ciudad']
df_restaurantes = df_restaurantes[['id_restaurante', 'id_ciudad', 'identificador_yelp', 'identificador_google', 'nombre', 'direccion', 'latitud', 'longitud', 'avg_rating', 'reviews_count', 'estado']]
```

```python
df_restaurantes.head(3)
```

```python
# Queremos saber que restaurantes estan abiertos, cuales cerrados temporalmente y cuales cerrados para siempre.
textos_a_mantener = ['Permanently closed', 'Temporarily closed']

# Identificar y reemplazar si es necesario
df_restaurantes['estado'] = df_restaurantes['estado'].apply(lambda x: 'Normally open' if x not in textos_a_mantener else x)
```
