```python
import pandas as pd
```

## Estandarizamos CATEGORIAS

```python
# Reemplaza 'nombre_del_archivo.csv' con el nombre de tu archivo CSV
nombre_archivo = 'categorias.csv'

# Carga el archivo CSV en un DataFrame
dfcategorias = pd.read_csv(nombre_archivo)
```

```python
dfcategorias
```

```python
# Eliminamos columnas que estan de mas.
dfcategorias = dfcategorias.drop(['Unnamed: 0'], axis=1)

# Para evitar problemas con los nombre, quitamos todos los ; de los datos y lo colocamos de separador.
dfcategorias = dfcategorias.applymap(lambda x: str(x).replace(';', ''))
dfcategorias.to_csv('dfcategorias.csv', sep=';', index=False)
```
