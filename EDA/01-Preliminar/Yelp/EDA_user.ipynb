### Vemos si tenemos duplicados en el user_id

```python
# Contar la cantidad total de duplicados en la columna 'business_id'
cantidad_duplicados = df['user_id'].duplicated().sum()

# Mostrar la cantidad total de duplicados
print("Cantidad total de duplicados en la columna 'user_id':", cantidad_duplicados)
```

### Eliminamos duplicados

```python
df = df.drop_duplicates()
```

```python
df.shape
```

En total fueron eliminadas 117700 filas

### Hacemos un top 5 de usuarios con mas reseÃ±as

```python
# Utiliza nlargest para obtener el top 5
top_5_review = df.nlargest(5, 'review_count')
top_5_review
```

```python
print(top_5_review["average_stars"])
```

### Hacemos un top 5 de usuarios con mas fans

```python
# Utiliza nlargest para obtener el top 5
top_5_fans = df.nlargest(5, 'fans')
top_5_fans
```

### Vemos el promedio de los que mas fans tienen

```python
print(top_5_fans["average_stars"])
```
